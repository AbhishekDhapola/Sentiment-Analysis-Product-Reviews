```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the simulated reviews data from the Excel file
excel_file_path = 'Simulation_output.xlsx'  # Update with your actual file path
reviews_data = pd.read_excel(excel_file_path)
```

```python
# Visualize the distribution of sentiment across products
plt.figure(figsize=(12, 6))
sns.countplot(x='Product', hue='Sentiment', data=reviews_data)
plt.title('Sentiment Distribution Across Products')
plt.xlabel('Product')
plt.ylabel('Count')
plt.legend(title='Sentiment', labels=['Negative', 'Positive'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
print(reviews_data.columns)
```

```python
reviews_data['Review Length'] = reviews_data['Review'].apply(len)

plt.figure(figsize=(12, 6))
sns.histplot(data=reviews_data, x='Review Length', hue='Sentiment', multiple='stack', bins=30)
plt.title('Distribution of Review Lengths')
plt.xlabel('Review Length')
plt.ylabel('Count')
plt.legend(title='Sentiment', labels=['Negative', 'Positive'])
plt.tight_layout()
plt.show()
```
